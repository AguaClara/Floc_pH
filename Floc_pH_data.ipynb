```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import math 

def panda_tsv(f):
  return pd.read_csv(f, sep='\t',header=None)

def parse_data(df):
  return pd.DataFrame(df.iloc[6:,:].to_numpy(),columns=df.iloc[5,:].to_numpy())

def plot_ANC_data(data):
  x = data["Titrant Volume (ml)"].to_numpy(dtype=float)
  y = data["pH"].to_numpy(dtype=float)
  max_vol = x[-1]

  plt.figure()
  plt.scatter(x,y)
  plt.title("Titration Curve for determining the ANC of tap water")
  plt.xlim(0,math.ceil(max_vol))
  plt.ylim(0,14)

  plt.xlabel("HCl Volume (ml)")
  plt.ylabel("pH")
  
BNC = panda_tsv("bnc_try1__3_2_20.tsv")
ANC1 = panda_tsv("try1__2_26_20.tsv")
ANC2 = panda_tsv("try2__3_1_20.tsv")
ANC3 = panda_tsv("try3__3_1_20.tsv")

df_arr = [BNC,ANC1,ANC2,ANC3]
data_arr = [parse_data(df) for df in df_arr]


for data in data_arr:
  plot_data(data)


```
